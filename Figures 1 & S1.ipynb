```python
''' Plot Figure 1D (Lick Rasters from Representative Mice [see percentiles further down]) '''

fig1d_mice = ['Y25','MA8F','A18']
idxs = [5,6,3,4,1,2]
seshnum = [1,6,1,6,1,6]

count = 0 

fig = plt.figure(figsize = (0.75*len(session),2.25))
nrow = 1
ncol = 6
dt = 0.02

tstarta = 20
tenda = 79
tstartb = 80
tendb = 139
tstartc = 140

nclick_ct = []
lick_ct = []

for m, session in zip(mice, sessions):
    
    if m not in fig1d_mice:
        continue 
    
    for s in session:
        
        if s in session[1:-1]:
            continue
        
        #get behavior data
        d = data[m][s]
        raw = d['raw']
        posx = raw.item().get('posx')
        post = raw.item().get('post')
        trial = get.trial_idx(posx)
        reward = raw.item().get('reward')
        trialinfo = raw.item().get('trialinfo')
        left = trialinfo['left']
        
        reward['trials'] = np.arange(0,len(reward['trials']),1)
        trials_dark = np.sum(raw.item().get('trialinfo')['dark'])
        lick_ct = d['reward_data']['lick_ct'].astype(bool)
        
        # get reward params & remove reward data pertaining to dark trials & skipped trials
        try:
            _ = raw.item().get('correctedtrial').shape
            trial = raw.item().get('correctedtrial')
            
            if (len(np.unique(trial)) < 220): 
                alltrial = np.arange(0,np.max(np.unique(trial)) + 1 , 1)
            else:    
                alltrial = np.arange(0,220,1)
                
            skippedtrials = get.find_skipped(trial, max_trial_num = 220)
            nonskipped = np.setdiff1d(reward['trials'],skippedtrials)   
            k_idx = np.intersect1d(np.where(reward['trials'] > trials_dark), nonskipped)
            keep_idx = np.intersect1d(np.where(reward['trials'] <= np.max(alltrial)), k_idx)
            
        except AttributeError:
            trial = raw.item().get('trial')
            keep_idx = (np.where(reward['trials'] > trials_dark))
            
        centers = np.array(reward['centers'][keep_idx])
        trials = np.array(reward['trials'][keep_idx])
        
        # remove any consummatory licks, defined as within the reward zone after first reward zone lick
        start_idx = []
        end_idx = []
        for t, c in zip(trials,centers):
            zone_start = (c - 25) #position of beginning of reward zone
            zone_end = (c + 25) #position of end of reward zone
            trial_posx = posx.copy()
            start_trial_ind = np.where(trial == t)[0][0]
            t_posx = trial_posx[trial == t]
            s_idx = np.argmin(np.abs(t_posx - zone_start)) 
            st_idx = np.argmin(np.abs(t_posx - zone_start)) + start_trial_ind
            e_idx = np.argmin(np.abs(t_posx[s_idx:] - zone_end)) + st_idx
    
            if np.sum(lick_ct[st_idx:e_idx]) > 0: 
                l1_idx = st_idx + np.nonzero(lick_ct[st_idx:e_idx])[0][0]
                start_idx.append(l1_idx)
                end_idx.append(e_idx)
            else:
                start_idx.append(st_idx)
                end_idx.append(e_idx)
                
        start_idx = np.asarray(start_idx)
        end_idx = np.asarray(end_idx)
        
        nclick_ct = lick_ct.copy()
        for i in range(start_idx.shape[0]):
            st = start_idx[i]
            e = end_idx[i] + 1
            nclick_ct[st:e] = False
        #print('Con Licks Removed:' + str(np.sum(lick_ct) - np.sum(nclick_ct)) + ', Non-Con Total: ' + str(np.sum(nclick_ct)))
        
        # sort lick train, posx by trial type only after trial 140
        right_long = [] #array of 0, 1 that identifies trial as context a vs. b; ensures context a is sorted 1st
        for t in trial:
            try: 
                if left[t] == 1:
                    right_long.append(0)
                else:
                    right_long.append(1)
            except IndexError:
                right_long.append(1)
        right_long = np.asarray(right_long)
                
        altidx = np.where(trial >= 140)[0][0]
        posx_tail = posx[altidx:]
        nclick_ct_tail = nclick_ct[altidx:]
        posx[altidx:] = posx_tail[np.argsort(right_long[altidx:], kind = 'stable')]
        nclick_ct[altidx:] = nclick_ct_tail[np.argsort(right_long[altidx:], kind = 'stable')]
        
        # get reward zone boundaries 
        rewarda = centers[20]
        rstarta = rewarda - 25
        renda = rewarda + 25
    
        rewardb = centers[80]
        rstartb = rewardb - 25
        rendb = rewardb + 25
        
        tendc = tstartc + np.sum(left[np.where(left[139:] == 1)])
        tstartd = tendc + 1
        tendd = tstartd + np.sum(left[np.where(left[139:] == 0)])
        
        #reverse everything
        trial = -1 * trial 
        
        #assign title color
        if m in SM_aged_mice:
            col = 'C1'
        elif m in SM_MA_mice:
            col = 'C2'
        else:
            col = 'C0'
        
        #make raster plots
        idx = idxs[count]
        ax = fig.add_subplot(nrow,ncol,idx)
        ax.plot(posx[nclick_ct], trial[nclick_ct], '.', fillstyle = 'full', markerfacecolor = 'k', markeredgecolor ='k', markersize=0.2, alpha = 0.4)
        ax.set_title(str(m) + '_' + str(seshnum[count]), color = col, fontsize = 10)
        ax.set_xlim([0, 400])
        ax.set_ylim([np.min(trial),0])
        ax.set_yticks([-220,-200,-180,-160,-140,-120,-100,-80,-60,-40,-20,0])
        ax.set_xticks([0,200,400])
        ax.tick_params(labelsize = 8)
        
        if idx == 1:
            ax.set_ylabel('Trials (n)', fontsize = 9)
            ax.set_yticklabels([220,200,180,160,140,120,100,80,60,40,20,0])
        elif idx == 4:
           
            ax.set_yticklabels([])
        else:
            ax.set_yticklabels([])
        
        #context a & b reward zone, c and d ranges are sorted alternation phase trials
        ax.fill_betweenx([-1*tstarta,-1*tenda], [rstarta, rstarta], [renda, renda], color='xkcd:pink', alpha = 1)
        ax.fill_betweenx([-1*tstartb,-1*tendb], [rstartb, rstartb], [rendb, rendb], color='xkcd:blue', alpha = 1)
        ax.fill_betweenx([-1*tstartc,-1*tendc], [rstarta, rstarta], [renda, renda], color='xkcd:pink', alpha = 1)
        ax.fill_betweenx([-1*tstartd,-1*tendd], [rstartb, rstartb], [rendb, rendb], color='xkcd:blue', alpha = 1)
                
        #indicate skipped trial reward zones with a light grey dashed line if they exist
        for i in skippedtrials:
            ax.hlines(i, 0,400, color = 'lightgrey', linewidth = 0.5, linestyles = 'dashed')
            
        count += 1

plt.subplots_adjust(hspace = 0.2, wspace = 0.4)
fig.supxlabel('VR Position (cm)', fontsize = 9, y = -0.05)
fig.savefig(save_folder + '1D.png', dpi = 400, bbox_inches='tight')
fig.savefig(save_folder + '1D.svg', dpi = 400, bbox_inches='tight')
plt.show()
```

```python
''' Calculate Total Non-Consummatory Licks / Trial & Frac. Licks in Opposite Zone / Trial '''

#session arrays
nclickb_rate = []
opplickb_rate = []
nclicka_rate = []
opplicka_rate = []

#animals arrays
mnclickb_rate = []
mopplickb_rate = []
mnclicka_rate = []
mopplicka_rate = []

for m, session in zip(mice,sessions):
    
    nclickb = []
    nclicka = []
    opplickb = [] 
    opplicka = [] 

    for s in session:
        
        #get behavioral data
        d = data[m][s]
        raw = d['raw']
        speed = raw.item().get('speed')
        posx = raw.item().get('posx')
        post = raw.item().get('post')
        lick_ct = d['reward_data']['lick_ct']
        trials_dark = np.sum(raw.item().get('trialinfo')['dark'])
        reward = raw.item().get('reward')
        reward['trials'] = np.arange(0,len(reward['trials']),1)
        dt = np.unique(np.round(np.diff(post),4))
        if len(dt) > 1:
            dt = dt[dt != 0]

        # remove reward data pertaining to dark trials & skipped trials
        try:
            _ = raw.item().get('correctedtrial').shape
            trial = raw.item().get('correctedtrial')
            
            if (len(np.unique(trial)) < 220): 
                alltrial = np.arange(0,np.max(np.unique(trial)) + 1 , 1)
            else:    
                alltrial = np.arange(0,220,1)
                
            skippedtrials = get.find_skipped(trial, max_trial_num = 220)
            nonskipped = np.setdiff1d(reward['trials'],skippedtrials)   
            k_idx = np.intersect1d(np.where(reward['trials'] > trials_dark), nonskipped)
            keep_idx = np.intersect1d(np.where(reward['trials'] <= np.max(alltrial)), k_idx)
            
        except AttributeError:
            trial = raw.item().get('trial')
            keep_idx = (np.where(reward['trials'] > trials_dark))
            skippedtrials = []
        
        centers = np.array(reward['centers'][keep_idx])
        trials = np.array(reward['trials'][keep_idx])
        misses = reward['missed'][keep_idx].astype(bool)
        requests = ~misses 

        # get indices for reward zone (center 25cm) & opp reward zone
        start_idx = []
        oppstart_idx = []
        end_idx = [] 
        oppend_idx = []
        
        for t, c in zip(trials, centers):
            trial_pos = posx.copy()
            start_trial_ind = np.where(trial == t)[0][0]
            t_posx = trial_pos[trial == t]

            zone_start = c - 25 
            zone_end = c + 25
            
            if c == 270: 
                oppzone_start = 370 - 25
                oppzone_end = 370 + 25
            else: 
                oppzone_start = 270 - 25
                oppzone_end = 270 + 25
            
            s_idx = np.argmin(np.abs(t_posx - zone_start)) 
            st_idx = s_idx + start_trial_ind
            e_idx = np.argmin(np.abs(t_posx[s_idx:] - zone_end)) + st_idx
            
            start_idx.append(st_idx)
            end_idx.append(e_idx)
            
            sopp_idx = np.argmin(np.abs(t_posx - oppzone_start))
            oppst_idx = sopp_idx + start_trial_ind
            oppe_idx = np.argmin(np.abs(t_posx[sopp_idx:] - oppzone_end)) + oppst_idx
            
            oppstart_idx.append(oppst_idx)
            oppend_idx.append(oppe_idx)
        
        start_idx = np.asarray(start_idx)
        end_idx = np.asarray(end_idx)
        
        oppstart_idx = np.asarray(oppstart_idx)
        oppend_idx = np.asarray(oppend_idx)
        
        # Remove any consummatory licks 
        nclick_ct = lick_ct.copy()
        for i in range(start_idx.shape[0]):
            st = start_idx[i]
            e = end_idx[i] + 1
            nclick_ct[st:e] = False
        
        # Get total nclicks/trial, save block & alternation mean for sessions, collect all values for animals
        ncbrate = []
        ncarate = []
        ncall = []
        
        for t in trials:
            ncall = np.append(ncall, np.sum(nclick_ct[trial == t]))
            if t < 140: 
                ncbrate.append(np.sum(nclick_ct[trial == t]))
            else: 
                ncarate.append(np.sum(nclick_ct[trial == t])) 
        
        nclickb = np.append(nclickb, ncbrate)
        nclicka = np.append(nclickb, ncarate)
        
        nclickb_rate = np.append(nclickb_rate, np.mean(ncbrate))
        nclicka_rate = np.append(nclicka_rate, np.mean(ncarate))
        
        # Get opposite reward zone licks, save block & alternation mean for sessions, collect all values for animals
        # Normalized to total non-con licks on that same trial
        oppbrate = []
        opparate = []
        for i, t in enumerate(trials):
            st = oppstart_idx[i]
            e = oppend_idx[i] + 1
            if t < 140:
                if np.sum(nclick_ct[trial == t]) > 0: 
                    oppbrate.append(np.sum(nclick_ct[st:e]) / np.sum(nclick_ct[trial == t]))
                else:
                    oppbrate.append(0)
            else:
                if np.sum(nclick_ct[trial == t]) > 0: 
                    opparate.append(np.sum(nclick_ct[st:e]) / np.sum(nclick_ct[trial == t]))
                else:
                    opparate.append(0)
        
        opplickb = np.append(opplickb, oppbrate)
        opplicka = np.append(opplickb, opparate)
        
        opplickb_rate = np.append(opplickb_rate, np.mean(oppbrate))
        opplicka_rate = np.append(opplicka_rate, np.mean(opparate))
              
    # get animal arrays
    mnclickb_rate = np.append(mnclickb_rate, np.mean(nclickb))
    mnclicka_rate = np.append(mnclicka_rate, np.mean(nclicka))
    
    mopplickb_rate = np.append(mopplickb_rate, np.mean(opplickb))
    mopplicka_rate = np.append(mopplicka_rate, np.mean(opplicka))
    
# ensure all outputs are saved as arrays
nclickb_rate = np.asarray(nclickb_rate)
nclicka_rate = np.asarray(nclicka_rate)
opplickb_rate = np.asarray(opplickb_rate)
opplicka_rate = np.asarray(opplicka_rate)

mnclickb_rate = np.asarray(mnclickb_rate)
mnclicka_rate = np.asarray(mnclicka_rate)
mopplickb_rate = np.asarray(mopplickb_rate)
mopplicka_rate = np.asarray(mopplicka_rate)
```

```python
''' Calculate & Save Lick N In- vs. Out-zone, Missed, Requested Rewards for Animals & Session'''
mlick_reward = []
mlick_out = []
mmissed_reward = []

lick_reward = []
lick_out = []
missed_reward = []

for m, session in zip(mice, sessions):
    for s in session:
        d = data[m][s]
        raw = d['raw']
        
        # get behavioral params
        posx = raw.item().get('posx')
        post = raw.item().get('post')
        trial = get.trial_idx(posx)
        if np.min(trial) > 0:
            trial = trial-1
        dt = np.unique(np.round(np.diff(post),4))
        speed = raw.item().get('speed')
        
        # get reward params
        lickt = raw.item().get('lickt')
        
        if m in cohortb_mice: 
            trials_dark = np.sum(raw.item().get('trial_dark'))
            reward = raw.item().get('reward')
            
            try:
                _ = raw.item().get('correctedtrial').shape
                trial = raw.item().get('correctedtrial')

                if (len(np.unique(trial)) < 240): 
                    alltrial = np.arange(0,np.max(np.unique(trial)) + 1 , 1)
                else:    
                    alltrial = np.arange(0,240,1)

                skippedtrials = np.setdiff1d(alltrial,np.unique(trial))
                nonskipped = np.setdiff1d(reward['trials'],skippedtrials)  
                nonskippedidx = []
                for s in nonskipped:
                    nonskippedidx = np.append(nonskippedidx, np.where(reward['trials'] == s)[0])
                k_idx = np.intersect1d(np.where(reward['trials'] > trials_dark), nonskippedidx)
                keep_idx = np.intersect1d(np.where(reward['trials'] <= np.max(alltrial)), k_idx).astype(int)
                #print('removed dark & skipped reward trials')
            
            except AttributeError:
                trial = raw.item().get('trial')
                k_idx = np.where(reward['trials'] > trials_dark)[0]
                keep_idx = np.intersect1d(np.where(reward['trials'] <= np.max(trial)), k_idx).astype(int)
                skippedtrials = []
                #print('removed dark reward trials')
                
            reward_centers = np.array(reward['centers'][keep_idx])
            reward_trials = np.array(reward['trials'][keep_idx])
        
        else: 
            trial = raw.item().get('trial')
            reward = raw.item().get('reward')
            reward_centers = np.array(reward['centers'])
            reward_trials = np.array(reward['trials'])
 
        lick_ct = d['reward_data']['lick_ct']
        
        # get indices for reward zone edges
        start_idx = []
        end_idx = []
        for t, c in zip(reward_trials, reward_centers):
            zone_start = c - 25
            zone_end = c + 25
            trial_pos = posx.copy()
            trial_pos[trial != t] = 1000
            start_idx.append(np.argmin(np.abs(trial_pos - zone_start)))
            end_idx.append(np.argmin(np.abs(trial_pos - zone_end)))
        start_idx = np.asarray(start_idx)
        end_idx = np.asarray(end_idx)
        
        # get number of licks per reward zone
        in_zone_licks = np.zeros_like(start_idx)
        missed = np.zeros_like(start_idx)
        
        for i in range(start_idx.shape[0]):
            s = start_idx[i]
            e = end_idx[i] + 1
            in_zone_licks[i] = np.sum(lick_ct[s:e])

        # get number of licks per trial
        total_licks = np.zeros_like(np.unique(trial))
        for i, t in enumerate(np.unique(trial)):
            total_licks[i] = np.sum(lick_ct[trial==t])
            
        # get N in/out zone licks by reward trial & save boolean for missed trials
        in_zone = np.asarray([])
        out_zone = np.asarray([])
        missed_zone = np.asarray([])
        for i, t in enumerate(np.unique(trial)):
            if t in reward_trials:
                
                total = total_licks[i]
                
                if total != 0:
                    licks_in = np.sum(in_zone_licks[np.where(reward_trials==t)[0]])
                    
                    if licks_in > total: #Corrects for error with negative out zone lick counts
                        in_zone = np.append(in_zone, 0)
                        out_zone = np.append(out_zone, 0)
                        missed_zone = np.append(missed_zone, False)
                    else: 
                        in_zone = np.append(in_zone, licks_in)
                        out_zone = np.append(out_zone, (total - licks_in))
                        missed_zone = np.append(missed_zone, False)
                else: 
                    in_zone = np.append(in_zone, 0)
                    out_zone = np.append(out_zone, 0)
                    missed_zone = np.append(missed_zone, True)
                    
        # save licks in zone and out of zone by session
        lick_reward.append(in_zone) #nested lists for each session and each mouse 
            #containing number of in zone licks for each reward trial
        lick_out.append(out_zone) #nested lists for each session and each mouse 
            #containing number of out zone licks during each reward trial
        missed_reward.append(missed_zone) #nested lists for each session and each mouse 
            #containing boolean of missed or not for each reward trial

    # get number of sessions for mouse
    n_session = len(session)
    size_lr = len(lick_reward)
    ref_lr = size_lr - n_session
    size_lo = len(lick_out)
    ref_lo = size_lo - n_session
    size_mr = len(missed_reward)
    ref_mr = size_mr - n_session
    
    # store lick in and out of zone across all trials & sessions by animal
    mlickreward = list(np.concatenate(lick_reward[ref_lr:]).flat)
    mlickout = list(np.concatenate(lick_out[ref_lo:]).flat)
    mmissedreward = list(np.concatenate(missed_reward[ref_mr:]).flat)
    
    # save mouse avg, sem speed, & mouse licks in & out 
    mlick_reward.append(mlickreward)
    mlick_out.append(mlickout)
    mmissed_reward.append(mmissedreward)
```

```python
''' Calculate RTA Speed & Licking for all Sessions, Centered to Reward (Plot Examples from Young & Aged Mice)'''
rewardspeed = np.zeros((1,30)) 
rewardlick = np.zeros((1,30)) 

len_rs = []
added_rs = []
len_rl = []
added_rl = []

loc_trough = []
trough_prom = []
loc_peak = []
peak_prom = []

from scipy.signal import peak_prominences as pp 

example_mice = ['A3','Y4']

for m, session in zip(mice,sessions):
    for s in session:
        d = data[m][s]

        raw = d['raw']
        speed = raw.item().get('speed')
        posx = raw.item().get('posx')
        trial = get.trial_idx(posx)
        if np.min(trial) > 0:
            trial = trial-1
        post = raw.item().get('post')
        dt = np.unique(np.round(np.diff(post),4))
        if len(dt) > 1:
            dt = dt[dt != 0]

        if m in cohortb_mice: #removes reward data pertaining to dark trials
            trials_dark = np.sum(raw.item().get('trial_dark'))
            reward = raw.item().get('reward')
            keep_idx = np.where(reward['trials'] > trials_dark)
            reward_centers = np.array(reward['centers'][keep_idx])
            rewardt = reward['times'][keep_idx]
            reward_trials = reward['trials'][keep_idx]

        else: 
            reward = raw.item().get('reward')
            reward_centers = np.array(reward['centers'])
            reward_trials = reward['trials']

        lick_ct = d['reward_data']['lick_ct']

        #get indices of peri-reward zone observations (25 cm before start + reward zone)
        start_idx = []
        center_idx = [] 
        end_idx = []
        for t, c in zip(reward_trials, reward_centers):
            zone_start = c - 50
            zone_end = c + 25
            trial_pos = posx.copy()
            trial_pos[trial != t] = 1000
            start_idx.append(np.argmin(np.abs(trial_pos - zone_start)))
            center_idx.append(np.argmin(np.abs(trial_pos - c )))
            end_idx.append(np.argmin(np.abs(trial_pos - zone_end)))
        start_idx = np.asarray(start_idx)
        center_idx = np.asarray(center_idx)
        end_idx = np.asarray(end_idx)

        # get running speed & lick counts pre- & in reward zones only
        in_zone_speed = []
        lick_ct_zone = []
        for i in range(start_idx.shape[0]):
            st = start_idx[i]
            e = end_idx[i] + 1
            in_zone_speed.append(speed[st:e])
            lick_ct_zone.append(lick_ct[st:e])

        #for each reward trial, get lick counts & speed binned by position peri-reward
        b = 2.5 # cm positions bins x 30 covering 75 cm (25 cm before + 50 cm reward zone)
        lick_ct_bin = np.zeros([reward_trials.shape[0],30]) #number of bins per reward zone differs
        speed_bin = np.zeros([reward_trials.shape[0],30])
        for i, t in enumerate(reward_trials):
            st = start_idx[i]
            e = end_idx[i] + 1
            in_zone_posx = posx[st:e]

            edges = np.arange(posx[st], posx[e] + b, b)
            centers = (edges[:-1] + edges[1:])/2
            b_idx = np.digitize(posx[st:e], edges) # for every frame in zone, gives which bin it falls into 
            if np.min(b_idx) == 0 : # reassign last frames or so in order to get same number of bins 
                b_idx[b_idx == 0] = 1
            if np.max(b_idx) > 30: # reassign last frames or so in order to get same number of bins 
                b_idx[b_idx > 30] = 30
            unique_bdx = np.unique(b_idx)

            #average speed in each bin, find lick count / occupancy per bin
            for j in range(unique_bdx.shape[0]):
                idx1 = (b_idx == unique_bdx[j]) # boolean index of whether frames on this reward trial are in this bin

                if np.sum(idx1) == 0:
                    print('warning zero occupancy')
                    lick_ct_bin[i, j] = 0 
                    speed_bin[i, j] = 0
                else:    
                    occupancy = dt * np.sum(idx1)
                    lick_ct_bin[i, j] = np.sum(lick_ct_zone[i][idx1]) / occupancy
                    speed_bin[i, j] = np.mean(in_zone_speed[i][idx1])  

        # Get binned average speeds & licks around reward centers; save out speed troughs & lick peaks
        window = np.linspace(-25, 50, 30)
        RTA_speed = np.mean(speed_bin, axis=0)
        rewardspeed = np.r_[rewardspeed,speed_bin]
        invRTA_speed = 0 - RTA_speed
        troughspeed = np.where(RTA_speed == np.min(RTA_speed))[0]
        troughprom = pp(invRTA_speed, troughspeed)[0]
        loctrough = window[troughspeed]

        RTA_licks = np.mean(lick_ct_bin, axis=0)
        rewardlick = np.r_[rewardlick,lick_ct_bin]
        peaklicks = np.where(RTA_licks == np.max(RTA_licks))[0]
        peakprom = pp(RTA_licks, peaklicks)[0]
        locpeak = window[peaklicks]
        
        loc_trough = np.append(loc_trough,loctrough)
        trough_prom = np.append(trough_prom,troughprom)
        loc_peak = np.append(loc_peak,locpeak)
        peak_prom = np.append(peak_prom,peakprom)

        # make plots
        if (m in example_mice) and ((s == session[0]) or (s == session[-1])):
            
            if m in RF_aged_mice:
                col = 'C1'
            else:
                col = 'C0'
            
            fig, (ax1,ax2) = plt.subplots(1, 2, sharex = True, figsize = (2.6,1.4))
            fig.tight_layout()
            fig.suptitle((str(m) + '_' + str(s[-1])), fontsize = 10, color = col)

            # plot speeds in reward zone and average
            for i in range(speed_bin.shape[0]):
                ax1.plot(window, speed_bin[i, :], '-k', linewidth=0.5, alpha=0.1)
            ax1.plot(window, RTA_speed, '-', linewidth=0.75, color = col, alpha=1)

            # format axes
            ax1.set_xlim([-25, 50])
            ax1.set_xticks([-25, 0, 25, 50])
            ax1.set_ylim([0, 75])
            ax1.set_yticks([0, 25, 50, 75])
            ax1.tick_params(labelsize = 8)
            ax1.set_ylabel('speed (cm / s)', fontsize = 9, labelpad = 1)

            # plot reward location
            y_lims = ax1.get_ylim()
            ax1.plot([0, 0], y_lims, '--k', linewidth=0.75)

            # plot licks in reward zone and average
            for i in range(lick_ct_bin.shape[0]):
                ax2.plot(window, lick_ct_bin[i, :], '-k', linewidth=0.5, alpha=0.1)
            ax2.plot(window, RTA_licks, '-', color=col, linewidth=0.75, alpha=1)

            # format axes
            ax2.set_xlim([-25, 50])
            ax2.set_xticks([-25, 0, 25, 50])
            ax2.set_ylim([0, 25])
            ax2.set_yticks([0, 10, 20])
            ax2.set_ylabel('licks', fontsize = 9, labelpad = 1)
            ax2.tick_params(labelsize = 8)

            # plot reward location
            y_lims = ax2.get_ylim()
            ax2.plot([0,0], y_lims, '--k', linewidth=0.75)

            fig.supxlabel('cm from reward start', fontsize = 9)
            plt.show()
    
    #Animal RTA Speed & Licks
    #Save number of new reward trials per animal
    len_rs = np.append(len_rs,len(rewardspeed) - 1)
    len_rl = np.append(len_rl,len(rewardlick) - 1)

added_rs = len_rs[0]
for m in range(1,len(len_rs)):
    added_rs = np.append(added_rs, len_rs[m] - len_rs[m-1])
    
added_rl = len_rl[0]
for m in range(1,len(len_rl)):
    added_rl = np.append(added_rl, len_rl[m] - len_rl[m-1])
```
